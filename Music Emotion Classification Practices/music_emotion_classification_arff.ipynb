# standardize the acoustic features and split the dataset for training and testing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the CSV file
df = pd.read_csv(df_path)

# Separate features and labels
X = df.iloc[:, :-6]  # All columns except last 6 (features)
y = df.iloc[:, -6:]  # Last 6 columns (emotion labels)

# Normalize features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Display dataset shapes
print("Training set:", X_train.shape, y_train.shape)
print("Testing set:", X_test.shape, y_test.shape)
```

  * 6 columns contain integer values (emotion labels).

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

model = Sequential([
    Dense(256, activation='relu', input_shape=(72,)),  # Input layer with 72 features
    Dropout(0.3),
    Dense(128, activation='relu'),  # First hidden layer
    Dropout(0.3),
    Dense(64, activation='relu'),  # Second hidden layer
    Dropout(0.3),
    Dense(32, activation='relu'),  # Third hidden layer
    Dropout(0.3),
    Dense(6, activation='softmax')  # Output layer with 6 emotion labels (sigmoid for multi-label)
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ------------------------------
# 5. Train the Model
# ------------------------------
history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                    epochs=100, batch_size=32)

# Evaluate on the test set
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy: {:.2f}%".format(accuracy * 100))
```

# FNN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the CSV file
df_path = '/content/drive/MyDrive/Thesis-Project/arff/emotions.csv'
df = pd.read_csv(df_path)

# Separate features and labels
X = df.iloc[:, :-6]  # All columns except last 6 (features)
y = df.iloc[:, -6:]  # Last 6 columns (emotion labels)

# Normalize features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the FNN model
fnn_model = Sequential([
    Dense(256, activation='relu', input_shape=(72,)),  # Input layer with 72 features
    Dropout(0.3),
    Dense(128, activation='relu'),  # First hidden layer
    Dropout(0.3),
    Dense(64, activation='relu'),  # Second hidden layer
    Dropout(0.3),
    Dense(32, activation='relu'),  # Third hidden layer
    Dropout(0.3),
    # Dense(6, activation='sigmoid')  # Output layer with 6 emotion labels (sigmoid for multi-label)
    Dense(6, activation='softmax')
])

# Compile the FNN model
fnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the FNN model
fnn_history = fnn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=32)

# Evaluate the FNN model
_, fnn_accuracy = fnn_model.evaluate(X_test, y_test)
print("FNN Test Accuracy: {:.2f}%".format(fnn_accuracy * 100))
```

```python
# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(fnn_history.history['accuracy'], label='Train Accuracy')
plt.plot(fnn_history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy')

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(fnn_history.history['loss'], label='Train Loss')
plt.plot(fnn_history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Reshape data for RNN (samples, timesteps, features) - assuming timesteps = 1
X_train_rnn = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test_rnn = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

# Define the RNN model
rnn_model = Sequential([
    LSTM(128, activation='relu', input_shape=(X_train_rnn.shape[1], X_train_rnn.shape[2])),  # LSTM layer
    Dropout(0.3),
    Dense(64, activation='relu'),  # Hidden layer
    Dropout(0.3),
    Dense(6, activation='sigmoid')  # Output layer with 6 emotion labels (sigmoid for multi-label)
])

# Compile the RNN model
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the RNN model
rnn_history = rnn_model.fit(X_train_rnn, y_train, validation_data=(X_test_rnn, y_test), epochs=100, batch_size=32)

# Evaluate the RNN model
_, rnn_accuracy = rnn_model.evaluate(X_test_rnn, y_test)
print("RNN Test Accuracy: {:.2f}%".format(rnn_accuracy * 100))
```

```python
import tensorflow.keras.backend as K

def f1_score(y_true, y_pred):
    y_true = K.cast(y_true, "float32")  # Ensure both are float32
    y_pred = K.round(y_pred)  # Convert predictions to binary (0 or 1)

    tp = K.sum(y_true * y_pred, axis=0)
    precision = tp / (K.sum(y_pred, axis=0) + K.epsilon())
    recall = tp / (K.sum(y_true, axis=0) + K.epsilon())

    f1 = 2 * (precision * recall) / (precision + recall + K.epsilon())
    return K.mean(f1)  # Average F1-score across all labels


model.compile(optimizer="adam", loss="binary_crossentropy", metrics=[f1_score])

from tensorflow.keras.callbacks import EarlyStopping

# Define early stopping callback
early_stopping = EarlyStopping(
    monitor="val_f1_score",  # Monitor validation F1-score
    patience=5,  # Stop if no improvement for 5 epochs
    mode="max",  # Maximize F1-score
    restore_best_weights=True  # Restore the best model
)

# Train the model with early stopping
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping],  # Add early stopping
    verbose=1
)

test_loss, test_f1 = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}, Test F1-Score: {test_f1}")
```

```python
import matplotlib.pyplot as plt

# Extract F1-score from history
train_f1 = history.history['f1_score']
val_f1 = history.history['val_f1_score']

# Plot F1-score
plt.figure(figsize=(8, 5))
plt.plot(train_f1, label="Train F1-score", marker='o')
plt.plot(val_f1, label="Validation F1-score", marker='o')
plt.xlabel("Epochs")
plt.ylabel("F1-score")
plt.title("F1-Score During Training")
plt.legend()
plt.grid()
plt.show()
```

# Hybrid Model (RNN + FNN) for Music Emotion Classification

```python
# Define the hybrid RNN + FNN model
hybrid_model = keras.Sequential([
    # RNN part
    layers.Input(shape=(72, 1)),
    layers.LSTM(128, return_sequences=True),
    layers.LSTM(64),

    # FNN part
    layers.Dense(128, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(64, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    # Output layer
    layers.Dense(6, activation="sigmoid")  # Multi-label classification
])

# Compile the model
hybrid_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Reshape input for LSTM
X_train_hybrid = X_train.reshape(-1, 72, 1)
X_test_hybrid = X_test.reshape(-1, 72, 1)

# Train the hybrid model
history_hybrid = hybrid_model.fit(
    X_train_hybrid, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test_hybrid, y_test),
    callbacks=[early_stopping]
)

# Evaluate the model on the test set
test_loss, test_accuracy = hybrid_model.evaluate(X_test_hybrid, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
